```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Task 1: Create 4 lines with different colors and styles

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.cos(2 * x)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, color='blue', linestyle='solid', label='sin(x)')
plt.plot(x, y2, color='red', linestyle='dashed', label='cos(x)')
plt.plot(x, y3, color='green', linestyle='dashdot', label='sin(2x)')
plt.plot(x, y4, color='purple', linestyle='dotted', label='cos(2x)')

plt.legend()
plt.title('Four Lines with Different Colors and Styles')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Task 2: Create a histogram with color bar
data = np.random.randn(1000)
plt.figure(figsize=(8, 6))

hist, bins,_= plt.hist(data, bins=20, alpha=0.7)

bin_centers = 0.5 * (bins[1:] + bins[:-1])  # Calculate bin centers

# Create a dummy scatter plot for color mapping
dummy = plt.scatter(bin_centers, hist, c=hist, cmap='viridis')

# Add a color bar to the plot
color_bar = plt.colorbar(dummy)
color_bar.set_label('Frequency')

plt.title('Histogram with Color Bar')
plt.xlabel('Value')
plt.ylabel('Frequency')
```

```python

# Task 3: Create hexagonal binnings
x_hex = np.random.randn(500)
y_hex = np.random.randn(500)
plt.figure(figsize=(8, 6))

plt.hexbin(x_hex, y_hex, gridsize=20, cmap='Blues')
plt.colorbar(label='Frequency')
plt.title('Hexagonal Binnings')
plt.xlabel('X')
plt.ylabel('Y')
```

```python

# Task 4: Create data with linspace and use scatter plots
x_scatter = np.linspace(0, 10, 50)
y_scatter = x_scatter + np.random.normal(scale=0.5, size=len(x_scatter))
plt.figure(figsize=(8, 6))

plt.scatter(x_scatter, y_scatter, color='blue', marker='o', label='Data Points')
plt.title('Scatter Plot of Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

```python

# Task 5: Create scatter plot with random size and colors and color bar
x_rand = np.random.rand(50)
y_rand = np.random.rand(50)
sizes = np.random.randint(20, 200, 50)
colors = np.random.rand(50)
plt.figure(figsize=(8, 6))

# cmap='viridis': This parameter specifies the colormap to use for mapping the colors parameter.
# The 'viridis' colormap is a perceptually uniform colormap that's often used for visualizations.

sc = plt.scatter(x_rand, y_rand, c=colors, s=sizes,cmap='viridis')
plt.colorbar(sc, label='Color Value')
plt.title('Scatter Plot with Random Size and Colors')
plt.xlabel('X')
plt.ylabel('Y')

```

```python
# Task 6: Create a figure with 3 subplots
plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.plot(x, y1, color='blue')
plt.title('Subplot 1')

plt.subplot(132)
plt.plot(x, y2, color='red')
plt.title('Subplot 2')

plt.subplot(133)
plt.plot(x, y3, color='green')
plt.title('Subplot 3')


# The plt.tight_layout() function is a utility function provided by Matplotlib that adjusts the spacing between subplots to prevent overlapping of plot elements. 
# It is typically used when you have multiple subplots in a figure and you want to make sure that the labels, titles, and other elements don't overlap with each other.
plt.tight_layout()

plt.show()
```
